```python
import pandas as pd
```

```python
il=pd.read_csv("E:/datasets/Insurance_suscribe.csv")
```

```python
il.head()
```

```python
il=il.iloc[:,1:]
```

```python
il.head()
```

```python
il.isnull().sum()
```

```python
il.describe(include='all')
```

```python
il=il.rename(columns={"Health Indicator":"Health_indicator"})
il.info()
```

```python
il.Health_indicator=il.Health_indicator.fillna('X1')
il.Holding_Policy_Duration=il.Holding_Policy_Duration.fillna('1.0')
il.Holding_Policy_Type=il.Holding_Policy_Type.fillna(il.Holding_Policy_Type.median())
```

```python
il.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
il[il.select_dtypes(include=['object']).columns]=il[il.select_dtypes(include=['object']).columns].apply(le.fit_transform)
```

```python
il.head()
```

```python
from sklearn.model_selection import train_test_split
il_train,il_test=train_test_split(il,test_size=0.2)
```

```python
train_x=il_train.iloc[:,0:-1]
train_y=il_train.iloc[:,-1]
test_x=il_test.iloc[:,0:-1]
test_y=il_test.iloc[:,-1]
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier( criterion='entropy',class_weight='balanced')
```

```python
dt.fit(train_x,train_y)
```

```python
pred=dt.predict(test_x)
```

```python
from sklearn.metrics import confusion_matrix
tab1=confusion_matrix(test_y,pred)
tab1
```

```python
dt.score(test_x,test_y)
```

```python
tab1.diagonal().sum()/tab1.sum()
```

```python
dt.feature_importances_
```

```python
Feature_importance=pd.DataFrame({"Feature":train_x.columns,"Importance":dt.feature_importances_})
Feature_importance
```
