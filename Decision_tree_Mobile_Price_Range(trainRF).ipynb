```python
from sklearn.model_selection import train_test_split
tr_train,tr_test=train_test_split(tr,test_size=0.2)
```

```python
train_x=tr_train.iloc[:,0:-1]
train_y=tr_train.iloc[:,-1]
test_x=tr_test.iloc[:,0:-1]
test_y=tr_test.iloc[:,-1]
```

```python
tr.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion='entropy',max_depth=9,
    min_samples_split=50,class_weight='balanced')
```

```python
dt.fit(train_x,train_y)
```

```python
dt.predict(test_x)
```

```python
from sklearn.metrics import confusion_matrix
tab1=confusion_matrix(test_y,dt.predict(test_x))
tab1
```

```python
tab1.diagonal().sum()/tab1.sum()
```
